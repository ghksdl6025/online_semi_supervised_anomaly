```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load your CSV
#    Change the filepath and column names as needed.
df = pd.read_csv('./data/0.049_noise.csv')

# 2. Sort & group into traces
#    Assume columns: "case:concept:name" for case ID, "concept:name" for activity, and "time:timestamp".
df = df.sort_values(["Case ID", "Timestamp"])
grouped = df.groupby("Case ID")["Activity"].apply(list)

# 3. Compute max trace length
max_len = grouped.map(len).max()
max_len = 36
# 4. For each prefix length L, count distinct prefixes
prefix_counts = []
for L in range(1, max_len + 1):
    # build a set of the first L events for each trace
    prefixes = {tuple(trace[:L]) for trace in grouped}
    prefix_counts.append((L, len(prefixes)))

# 5. Put into DataFrame
results = pd.DataFrame(prefix_counts, columns=["prefix_length", "num_variants"])

# 6. Plot it
plt.figure(figsize=(8, 5))
plt.plot(results["prefix_length"], results["num_variants"], marker="o")
plt.title("Log Complexity by Prefix Length")
plt.xlabel("Prefix Length")
plt.ylabel("Number of Distinct Prefix Variants")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
results['Per prefix'] = round(results['num_variants']/results['prefix_length'],2)
```
